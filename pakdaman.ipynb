```python
import pandas as pd
import numpy as np
data = pd.read_csv('pakdaman.csv')
from sklearn.model_selection import train_test_split
X,y = data.iloc[:,:-1] , data.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error
# np.any(np.isnan(X))
# np.where(np.isnan(X)) 
# np.any(np.isnan(y))
# np.all(np.isfinite(X))
# np.all(np.isfinite(y))
abr_base = AdaBoostRegressor()
abr_base.fit(x_train,y_train)
abr_deci = AdaBoostRegressor(DecisionTreeRegressor())
abr_deci.fit(x_train,y_train)
abr_extr = AdaBoostRegressor(ExtraTreeRegressor())
abr_extr.fit(x_train,y_train)
```

```python
predict_base = abr_base.predict(x_test)
predict_deci = abr_deci.predict(x_test)
predict_extr = abr_extr.predict(x_test)
rmse_BaseEstimator = np.sqrt(mean_squared_error(y_test,predict_base))
rmse_DecisionTree = np.sqrt(mean_squared_error(y_test,predict_deci))
rmse_ExtraTree = np.sqrt(mean_squared_error(y_test,predict_extr))
print("RMSE_BaseEstimator: %f" % (rmse_BaseEstimator))
print("RMSE_DecisionTree: %f" % (rmse_DecisionTree))
print("RMSE_ExtraTree: %f" % (rmse_ExtraTree))
brmse2 = 100
for k in range(3, 20, 1):
    abr_deci = AdaBoostRegressor(DecisionTreeRegressor(max_depth=k))
    abr_deci.fit(x_train,y_train)
    predict_decis = abr_deci.predict(x_train)
    rmsemax = np.sqrt(mean_squared_error(y_test,predict_deci))
    print("max_depth = ", k)
    print("rmse = %f"  % (rmsemax))
    if rmsemax < brmse2 :
            brmse2 = rmse_BaseEstimator
            ko = k
print("best RMSE: %f" % (brmse2))
print("best k = ", ko)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('pakdaman.csv')
from sklearn.model_selection import train_test_split
X,y = data.iloc[:,:-1] , data.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error
lr = [0.1, 0.2, 0.5, 0.8, 1, 1.5, 2, 2.5, 3]
brmse = 100
for i in range(20, 300, 10):
    for j in lr:
        abr_base = AdaBoostRegressor(n_estimators=i, learning_rate=j)
        abr_base.fit(x_train,y_train)
        predict_base = abr_base.predict(x_test)
        rmse_BaseEstimator = np.sqrt(mean_squared_error(y_test,predict_base))
        print("i = " ,i)
        print("j = " ,j)
        print("RMSE_BaseEstimator: %f" % (rmse_BaseEstimator))
        if rmse_BaseEstimator < brmse :
            brmse = rmse_BaseEstimator
            io = i
            jo = j
print("best RMSE: %f" % (brmse))
print("best i = " ,io)
print("best j = " ,jo)
```
